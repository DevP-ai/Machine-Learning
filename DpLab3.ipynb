# **Normalization**

```python
placement=pd.read_csv("placement.csv")
```

```python
placement.sample(5)
```

## **Min-Max**

```python
min_max=placement.copy()
```

```python
new_min=2
new_max=6
for columns in min_max.columns:
  min_max[columns]=((min_max[columns]-min_max[columns].min()) /(min_max[columns].max()-min_max[columns].min()))*(new_max-new_min)+new_min
```

```python

min_max.sample(10)
```

## **Decimal Scaling**

```python
decimal=placement.copy()
```

```python
for column in decimal:
  decimal[column]=decimal[column]/decimal[column].abs().max()
```

```python
display(decimal)
```

## **Z-Score**

```python
z_score=placement.copy()
```

```python
for column in z_score:
  z_score[column]=(z_score[column]-z_score[column].mean())/z_score[column].std()
```
