```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("/content/wine_data.csv",header=None,usecols=[0,1,2])
df.columns=["Class","Alcohol","Malic Acid"]
```

```python
df.sample(5)
```

```python
sns.kdeplot(df['Alcohol'])

plt.show()
```

```python
sns.kdeplot(df['Malic Acid'])
```

```python
sns.kdeplot(df['Alcohol'])
sns.kdeplot(df['Malic Acid'])
plt.show()
```

```python
color_dic={1:'red',2:'blue',3:'green'}
sns.scatterplot(df['Alcohol'],df['Malic Acid'],hue=df['Class'],palette=color_dic)
plt.show()
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(df.drop('Class',axis=1),
                                               df['Class'],random_state=0,test_size=0.3)

X_train.shape,X_test.shape,y_train.shape,y_test.shape
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()
```

```python
scaler.fit(X_train)
```

```python
X_train_scaled=scaler.transform(X_train)
X_test_scaled=scaler.transform(X_test)
```

```python
X_train_scaled=pd.DataFrame(X_train_scaled,columns=X_train.columns)
X_test_scaled=pd.DataFrame(X_test_scaled,columns=X_test.columns)
```

```python
X_train_scaled.head(2)
```

```python
X_train.head(2)
```

```python
np.round(X_train.describe(),1)
```

```python
np.round(X_train_scaled.describe(),1)
```

```python
fig,(ax1,ax2)=plt.subplots(ncols=2,figsize=(12,5))

ax1.scatter(X_train['Alcohol'],X_train['Malic Acid'],c=y_train)
ax1.set_title("Before Mix-Max")

ax2.scatter(X_train_scaled['Alcohol'],X_train_scaled['Malic Acid'],c=y_train)
ax2.set_title("After MinMax")

plt.show()
```

```python
fig,(ax1,ax2)=plt.subplots(ncols=2,figsize=(12,5))

sns.kdeplot(X_train['Alcohol'],ax=ax1)
sns.kdeplot(X_train['Malic Acid'],ax=ax1)
ax1.set_title("Before Mix-Max")

sns.kdeplot(X_train_scaled['Alcohol'],ax=ax2)
sns.kdeplot(X_train_scaled['Malic Acid'],ax=ax2)
ax2.set_title("After MinMax")

plt.show()
```
