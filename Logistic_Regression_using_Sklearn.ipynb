```python
import numpy as np
import pandas as pd
```

```python
df=pd.read_excel("/content/PPG-BP dataset.xlsx",header=1)
```

```python
df.sample(5)
```

```python
df['Hypertension'].replace({'Normal':1},inplace=True)
df['Hypertension'].replace({'Prehypertension':1},inplace=True)
df['Hypertension'].replace({'Stage 1 hypertension':0},inplace=True)
df['Hypertension'].replace({'Stage 2 hypertension':0},inplace=True)
```

```python
df.sample(10)
```

```python
feature=["BMI(kg/m^2)","Heart Rate(b/m)","Diastolic Blood Pressure(mmHg)"]
```

```python
X=df[feature].values
```

```python
y=df["Hypertension"].values
```

```python
X
```

```python
X.shape
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25, random_state = 0)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
sc_x = StandardScaler()
xtrain = sc_x.fit_transform(X_train) 
xtest = sc_x.transform(X_test)
```

```python
from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression(random_state = 0)
classifier.fit(xtrain, y_train)
```

```python
y_pred = classifier.predict(xtest)
```

```python

from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_pred)
  
print ("Confusion Matrix : \n", cm)
```

```python

from sklearn.metrics import accuracy_score
print ("Accuracy : ", accuracy_score(y_test, y_pred))
```

```python
from sklearn.metrics import precision_recall_fscore_support
precision_recall_fscore_support(y_test, y_pred, average='macro') 
```

```python
precision_recall_fscore_support(y_test, y_pred, average='weighted')
```
